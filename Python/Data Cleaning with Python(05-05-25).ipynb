```python
# Import pandas and numpy libraries 
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer # to handle numerric data strategy is mean or median, for categorical data strategy is most frequent
```

```python
# load the data set
data = pd.read_csv(r'C:/Users/Admin/1  Bharat Digital Python files(Udaanous)/05-05-25/Messy data set.csv') # load excel file  use pd.read_excel("file path name")
data.head() # method to show 1st 5 rows
```

```python
missing_data = data.isnull().sum() # check if the values are null or missing 
print("Missing Data: ") 
missing_data
```

```python
# changing the data type of id,age,salary and error handeling 
data['ID'] = pd.to_numeric(data['ID'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')
data
```

```python
# to handle numerric data strategy is mean or median, for categorical data strategy is most frequent
imputer= SimpleImputer(strategy='mean')
imputer.fit(data[['Age','Salary']])
data[['Age','Salary']]= imputer.transform(data[['Age','Salary']])

data
```

```python
data.dtypes # methode which provide all the data types
```

```python
# remove rows with missing values
data.dropna(inplace=True)
data
```

```python
# assigning data types
data['ID']=data['ID'].astype(int)
data
```

```python
# storing unique values and removing inconsistent values
unique_values=data['ID'].unique() # function to find unique values
unique_values
```

```python
data=data[data['ID']!='inconsistent_value'] # filter the data again and takes all values except inconsistant values
data
```

```python
# Dealing with duplicates
duplicates=data.duplicated()
data[duplicates]
```

```python
# remove the duplicates
data.drop_duplicates(inplace=True) # inplace= True ensures the duplicates are removed from original data. 
data
```
