# Features Selecteion 

```python
#Plotting the correlation graph to see the correlation between features and dependent variable.
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
# Selecting the features using VIF. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list=[]
for col in df.columns:
    if ((df[col].dtype!='object') & (col!='price')):
        col_list.append(col)

x=df[col_list]
vif=pd.DataFrame()
vif['feature']=x.columns
vif['VIF']=[variance_inflation_factor(x.values,i) 
            for i in range(len(x.columns))]
```

```python
vif
```

 Selecting the features using
VIF. VIF should be less than 5. Feature that VIF value is greater than 5 drop this column. 

```python
col=['flight','price']
x=df.drop(columns=col)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# LinearRegression

```python
# Applying standardization and implementing Linear Regression Model to predict the price of a flight.
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)

difference=pd.DataFrame(np.c_[y_test,y_pred], columns= ["Actual_value","Predicted_value"])
difference
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np
```

```python
# R² Score
r2 = r2_score(y_test, y_pred)

# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Mean Absolute Percentage Error
mape = mean_absolute_percentage_error(y_test, y_pred)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# Print all metrics
print(f"R² Score: {r2}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

After Calculating r2 score,MAE, MAPE, MSE, RMSE. Root Mean square error(RMSE)
of the Linear regression model is 7014.30 and Mean absolute percentage
error(MAPE) is 43 percent. Lower the RMSE and MAPE better the model.

```python
# Plotting the graph of actual and predicted price of flight
sns.distplot(y_test,label='Actual')
sns.distplot(y_pred,label='Predicted')
plt.legend()
plt.show()
```
