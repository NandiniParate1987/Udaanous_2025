# Bar chart

```python
# Calculate frequency of churned values
freq_churned = data['Churn'].value_counts()

# Plot using matplotlib
plt.figure(figsize=(8, 5))
freq_churned.plot(kind='bar')
plt.title('Churned Customer Frequency')
plt.xlabel('Churned')
plt.ylabel('Count')
plt.show()

# Countplot using seaborn
sns.countplot(x='Gender', data=data)
plt.title('Category Distribution')
plt.show()

# Countplot using seaborn
plt.figure(figsize=(10, 20))
sns.countplot(x='Age', data=data)
plt.title('Category Distribution')
plt.show()

# Countplot using seaborn
plt.figure(figsize=(10, 20))
sns.countplot(x='Tenure', data=data)
plt.title('Category Distribution')
plt.show()
```

# stack and Clustered 

```python
data.columns = data.columns.str.strip()
# Create the crosstab
cross_tab = pd.crosstab(data['Gender'], data['Contract Length'])
plt.figure(figsize=(8, 5))
cross_tab.plot(kind='bar', stacked=False)
plt.title('Gender-wise Contract Length Distribution')
plt.xlabel('Gender')
plt.ylabel('Total Counts')
plt.show()
sns.countplot(x='Gender', hue='Contract Length', data=data)     

           
```

# Pie chart

```python

freq_churned = data['Subscription Type'].value_counts()
plt.figure(figsize=(8, 5))
freq_churned.plot(kind='pie', autopct='%1.2f%%')  # fixed quote here
plt.title('The title of the graph')
plt.ylabel('')  # Better to use empty string than a space
plt.show()
```

```python
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
freq_churned = data['Subscription Type'].value_counts()
plt.figure(figsize=(8, 5))
freq_churned.plot(kind='line', marker='o')
plt.title('Gender wise Total Spend Distribution')
plt.xlabel('Gender')
plt.ylabel('Subscription Type')
plt.show()
sns.lineplot(x='Subscription Type', y='Gender', data=data, marker='o')          
```
