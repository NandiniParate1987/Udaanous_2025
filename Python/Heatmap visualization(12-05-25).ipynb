```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data=pd.read_excel(r'C:/Users/Admin/1  Bharat Digital Python files(Udaanous)/12-05-25/heatmap_data.xlsx')
data.head(10)
```

```python
# Suppress all warnings
warnings.filterwarnings('ignore')

# Compute correlation matrix
correlation_matrix = data.corr().round(3)
print(correlation_matrix)

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True)
plt.title('The Correlation matrix')
plt.show()
```

```python

columns = ['Task Completion Rates', 'Decision-Making Skills Rating', 'Sales Revenue Generated']
data_selected = data[columns]
# Compute the correlation matrix
correlation_matrix = data_selected.corr().round(3)
print(correlation_matrix)

# Plot using matplotlib only
fig, ax = plt.subplots(figsize=(8, 5))
cax = ax.matshow(correlation_matrix, cmap='coolwarm')

# Add colorbar
fig.colorbar(cax)

# Set ticks and labels
ax.set_xticks(range(len(correlation_matrix.columns)))
ax.set_yticks(range(len(correlation_matrix.columns)))
ax.set_xticklabels(correlation_matrix.columns, rotation=45, ha='left')
ax.set_yticklabels(correlation_matrix.columns)

# Annotate correlation values
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        value = correlation_matrix.iloc[i, j]
        ax.text(j, i, str(value), va='center', ha='center', color='black')

plt.title('Correlation Matrix (Matplotlib Only)', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Plot the heatmap using seaborn with selected column
columns = ['Task Completion Rates', 'Decision-Making Skills Rating', 'Sales Revenue Generated']
data_selected = data[columns]

# Compute the correlation matrix
correlation_matrix = data_selected.corr().round(3)
print(correlation_matrix)

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='gray')
plt.title('Correlation Matrix of Performance Metrics')
plt.tight_layout()
plt.show()
```

Conclusion: All the three columns('Task Completion Rates', 'Decision-Making Skills Rating', 'Sales Revenue Generated') correlated with each other

 Task Completion Rates are highly correlated with Decision-Making Skills Rating with corr_matrix= 0.71
 
 Decision-Making Skills Rating are negatively correlated with Sales Revenue Generated with corr_matrix= -0.059
